It does not test all functionality nor serve as a tutorial.

```python
# Load a single simulation

from calpycles.pycles_sample import SampleDYCOMS_RF01

import os
this_path = os.path.abspath("")

nature_lowres = SampleDYCOMS_RF01(
    name="nature_lowres",
    parent_path=this_path,
    load=True,
)
```

```python
# Plot the simulation togehter with the measurement data

from calpycles.plotting.ens_obs import plot_ens_obs

from calpycles.DYCOMS_RF01.measurements import MeasurementsDYCOMS_RF01
meas = MeasurementsDYCOMS_RF01()

plot_ens_obs(
    ensembles=[],
    ens_labels=[],
    data=meas,
    nature=nature_lowres,
    nature_name="Nature (LOWRES)",
    save_file=f"fig-obs_nature_lowres",
)
```
