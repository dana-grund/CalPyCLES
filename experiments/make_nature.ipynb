```python
from calpycles.pycles_sample import SampleDYCOMS_RF01
from calpycles.parameters import ParametersDYCOMS_RF01 as Params
import os

from data_loaders import load_nature
from data_loaders import model_names
from data_loaders import make_namelist_settings

# location
this_path = os.path.abspath("")
data_folder = os.path.join(this_path,"data/")
name = "nature" # sample name, same for all
```

```python
# send samples to run in jobs

for i, model_name in enumerate(model_names):

    if i==0:
        continue 
    if i==1:
        continue 

    sample = SampleDYCOMS_RF01(
        name=name,
        parent_path=os.path.join(data_folder, model_name),
        namelist_settings=make_namelist_settings(model_name)
    )

    if model_name in [
        "MIXED_FULLRES",
        "CENTRAL_FULLRES",
    ]:
        # set nonzero cs defaults
        params = Params.defaults
        names = Params.names 
        i_cs = names.index("cs")
        params[i_cs] = 0.17
        sample.parameters.set_from_np(params)

    sample.namelist.write()

    sample.run(
        submit_job= True,
        do_wait_to_finish = False,  # cant do cleaning!
    )
```

```python
# after samples finished, prepare evaluation

for i, model_name in enumerate(model_names):
    print(model_name)
    nature = load_nature(model_name)
    nature.clean_files(**{
            # for nature, keep all information
            "delete_fields": False,
            "select_stats": False,
            "delete_cond_stats": False,
    })
    nature.observable.observe()
```
